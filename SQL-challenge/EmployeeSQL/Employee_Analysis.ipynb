```python
import pandas as pd

import matplotlib.pyplot as plt
from sqlalchemy import create_engine
import psycopg2

#Import password
from postgrespwd import pwd

salaries = pd.read_csv("../data/salaries.csv")
salaries.head()
```

```python
sorted_salaries = salaries.sort_values('salary', ascending = False)
sorted_salaries
```

```python
sorted_salaries.dtypes
```

```python
#Creating bins to group salaries:
bins = [0, 49999, 59999,69999,79999,89999,99999,109999,119999,129999]

# Create labels for these bins
salaries = ["<50000", "50000-59999", "60000-69999", "70000-79999", "80000-89999","90000-99999", "100000-109999", 
              "110000-119999","120000-129999"]

#Assign the data into the new age categories and create a new column into purchase_data_df:
salary_category = pd.cut(sorted_salaries["salary"], bins, labels=salaries, include_lowest= True )
sorted_salaries["salary_category"] = salary_category
sorted_salaries
```

```python
#Group by salary categories
categories = sorted_salaries.groupby("salary_category")
#Count the number of employees per category
emp_cat = categories["salary_category"].count()
emp_cat
```
